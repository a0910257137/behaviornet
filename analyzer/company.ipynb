```python
import numpy as np 
from matplotlib import pyplot as plt
from scipy.fft import fft, fftfreq, ifft
```

```python
path = "/aidata/anders/objects/company/meas_scope_20210823.txt"

def load_text(path):
    with open(path) as f:
        return [l.replace("\n", "") for l in f.readlines()]
text_lines = load_text(path)
```

```python
time, ch1, ch2 = [], [], []
for i, line in enumerate(text_lines):
    if i < 6:
        continue
    line = line.split("; ")
    
    try:
        time.append(float(line[0]))
        ch1.append(float(line[1]))
        ch2.append(float(line[2]))
    except:
        print("failed convert in {} line".format(i))
```

```python
fig = plt.figure()
time = np.asarray(time)
ch1 = np.asarray(ch1)
ch2 = np.asarray(ch2)
plt.plot(time, ch1, color = "orange", alpha= 0.5)
plt.grid()
plt.show()
```
